## Introduction

In this article/notebook, I'll discuss Mankiw's brief and prominent counterpoint (linked below)[^1] to Piketty's famous `r > g` equation from his influential work, *Capital in the Twenty-First Century*[^2]. Why a Python notebook? I'm not entirely sure, though I'll note that Mankiw's response answers Piketty math for math. Call it a hunch.

I should also mention that I was given the article, nearly ten years old, in the context of the study of complexity economics, an school of thought which Iʻll attempt to dip into here, if only just a bit. Complexity economics seems to have quite a bit to offer on the subject of dispute between Mankiw and Piketty; significantly more, it would seem, than Mankiw himself.

## Premise

Let us begin with a defense of the article. By 2015, America had not yet become a living laboratory for `r > g`, through its repeated election of a living embodiment of misanthropic dynastic wealth. (Some will say we had indeed experienced the very same by 2015, and that Professor Mankiw himself had had a front row seat. Clearly, though, there are demons lesser and greater.) This is about the only defense I can think of. I think elements of the article are understandable, when viewed through a certain lens; one that would perhaps be otherwise attached to a rifle of some kind.

Mankiw peers through that lens with both eyes early in his article: "To someone who views relatively unfettered capitalism as one of the great achievements of human history and the best way to organize a society, as I do..."[^1] ...and at once we are left to wonder if we are to read a scientific article or a stump speech.

(By the way, when you pick a "great achievement of human history," consider selecting one that needn't be double-qualified as "relatively unfettered." For instance, "The written word." "Women's rights." "Spaceflight." "Pre-natal care." Or, as a proscriptionary example: "Relatively unfettered pre-natal care.")

The article continues about as expected, though the next bit is surprising in degree:

> The first thing to say about Piketty's logic is that it will seem strange to any economist trained in the neoclassical theory of economic growth... we should be reassured that we live in a world in which r > g because it means we have not left any dynamic Pareto improvements unexploited.[^1]

In other words, "Of course r > g: the model wouldn't work correctly if it isn't that way." 

And quickly we see the problem whooshing over Mankiw's head: except that this should be impossible—*is* impossible—for a scholar with Professor Mankiw's towering qualifications. We're left to immediately wonder what could possibly be going on behind the scenes here. Obviously, if a "working" model is predicting the kinds of effects that lead to societal discord sufficient to put American Democracy, of late, on the ropes, we might consider ourselves chastened, rather than consoled, by its smooth linear functioning. 

Is there no `r > g` that Mankiw would find troubling? In *The Matrix*, the robot empire seemed to enjoy a solid rate of return on a rather poor growth rate for humanity. The machines were excellent Pareto optimizers. r > g, so what, Neo?[^3]

Next, we make our way to the Three Vexations of the Filthy Rich, or, as Mankiw puts it, "He can pass his wealth on to his children, but to ensure that his descendants remain wealthy, he faces three obstacles."[^1]

- **Consumption** ("and riotous living but also political and philanthropic contributions"[^1]): This paragraph seems almost designed to evoke a reaction from a particular set of readers, but from a math perspective: isn't consumption just a tiny intricate scalar on `r` at best? I say "at best," since I suspect that actual financial consumption among the most rich is shockingly low relative to lifestyle—after all, rent, for billionaires, often never comes due.

- **Inheritance**: Here, we've at least graduated to a factor on `r`, but as Mankiw glancingly acknowledges and then swiftly discards, money isn't the only thing being bequeathed from generation to generation through "procreation" (the choice of this descriptor being an interesting flourish). The idea that wealth dilutes cleanly through generations seems to ignore the rather obvious fact that inherited advantages compound rather than divide. I suspect there is something more fundamentally flawed about this argument related to discrete time and asynchronous inputs, but that idea deserves a more detailed reading than this one.

- **Taxes**: Again with a positive scalar. Unless you're sufficiently billionaire, of course, in which case the entire scalar `<= 0`. Itʻs tickling, in context, to note that taxes on capital gains are a critical factor that keeps Mankiwʻs entire formulation from flying apart; one wonders if Mankiw's fierce advocacy for the equilibrium necessity of circa-2015 tax rates ever cost him a dinner invitation. (Presumably not; this is *math* we are discussing.)

## Convenient Assumptions and Fairy-Tale Models

As Mankiw inventories the harsh realities facing the modern heir or heiress, he quantifies his work. Consumption, he calculates, is worth about 3 percent of rates of return, though directly quoting the sentence in which this "calculated" would be too unkind. Inheritance is tabbed as a sort of "doubling time" of about 35 years, downshifting the factor to a measly scalar, 2 percent as the crow flies. For taxes, we see:
>" In Massachusetts, where I live, the state imposes an additional estate tax with a top rate of 16 percent. As a result... estate taxation reduces the accumulation of dynastic wealth by about 2 percent per year."[^1]

Still no error bars, which would seem to be a problem if Mankiw's chief argument is `r - 7 <> g`. 

However: not the only problem! Apparently Mankiw's argument only works in Massachusetts (I am certain I have read somewhere about that state's tax policies), and, more challengingly, only works in Massachusetts in 2015, with 2015 consumption rates and 2015 population demographics. As we'll see, the approach depends on our scalars adding up fairly precisely to minus-seven. And all this is taking on faith—the article leaves us nothing else—the assumption that they ever added up anywhere, anytime, in the first place.

However, the most important flaw in the Mankiw comes in the next paragraph:

> "This [Piketty's] scenario is far from what we have experienced. Piketty estimates the real rate of return to be about 4 or 5 percent, which seems plausible for a typical balanced portfolio. Meanwhile, the average growth rate of the US economy has been about 3 percent. So Piketty is right that r has exceeded g, but it has done so by only about 2 percentage points, not the more than 7 percentage points necessary for the creation of Piketty's imagined dystopia."[^1]

How fortunate, then, for the very wealthy, that they do not have to suffer through the doldrums of a "typical balanced portfolio." The idea that the very wealthy do not enjoy far superior returns on capital than average seems so obvious that it is difficult to imagine how Mankiw could have, as an impartial analyst, overlooked it. Especially in light of the fact that Piketty devotes an entire chapter (12) to the topic, arriving at the understanding that:

> "Nevertheless, no matter what years we choose, the structural rate of growth of the largest fortunes seems always to be greater than the average growth of the average fortune (roughly at least twice as great)."[^2]

Even if Mankiw doesnʻt believe the evidence, or he didnʻt read that far in the Piketty, he had certainly been surrounded by this point with a community of like-minded individuals who would *almost never shut up* about their ability to command superior rates of return. Instead we are given `r - 7` and the lid is slipped deftly back onto the cookie jar.

The omission of this key point guts the logic of his article. If wealthy investors earn rates of return about double those enjoyed by retirement funds for schoolteachers (as Piketty suggests), then there is some `r*` for which `r* > r > g`. More precisely, if `r* = 2r` as Piketty indicates, and Mankiw's own estimate puts `r` at 4-5%, then `r*` would be 8-10%—well above the `r - 7` threshold that Mankiw claims is necessary for Piketty's "imagined dystopia." Again, this was back in the good olʻ days of 2015.

Maybe Mankiw is right, and Pikettyʻs `r` isnʻt a problem. We can comfortably assert, however, that `r*` is. In fact, Piketty does, and Mankiw doesnʻt mention it.

The meaningful balance of the article builds a petite theoretical model on these shifting and possibly illusory sands.

> Consider an economy composed of two kinds of people—workers and capitalists. ... Workers and capitalists come together to produce output, using a production function that experiences labor-augmenting technological progress, and they earn the value of their marginal product. In addition, following the advice of Piketty, the government imposes a tax on capital equal to τ per period, the revenue from which is transferred to workers.

From a model builder standpoint, what follows genuinely looks like a lark to try to building.[^4]

With entrepreneurs (presumably men) outfitted in their top hats and monocles, and entrepreneurs in their coveralls and signature lack of agency, we are given the system that follows.

Paraphrasing the narrative:

- `c_w` : consumption of each worker
- `c_k` : consumption of each capitalist
- `w` : wage
- `r` : before-tax rate of return on capital
- `κ` : capital stock per worker
- `n` : number of workers per capitalist
- `nκ` : capital stock per capitalist
- `f(κ)` : production function for output (net of depreciation)
- `g` : rate of labor-augmenting technological change (steady-state growth rate)
- `σ` : capitalists' intertemporal elasticity of substitution
- `ρ` : capitalists' rate of time preference
- `τ` : tax rate on capital per period

(Here, We use kappa to avoid confusion between capital stock and capitalist.)

...and then:

$$
\begin{align}
c_w &= w + \tau \kappa \\
c_k &= (r - \tau - g)n\kappa \\
r &= f'(\kappa) \\
w &= f(\kappa) - r\kappa \\
g &= \sigma(r - \tau - \rho)
\end{align}
$$

Again, a delightful toy, and one that certainly has been echoed in complexity simulations before and since. It all makes sense and looks familiar.

However, if there is just one thing[^5] that Complexity Economics is terrific at illustrating, it is that a model composed only of flows between capitalists and workers is a model that is empirically blind to real modern-day economic function. 

For instance, citing to a mainstay article from the field (and somewhat contemporary to Mankiw), Sawyer & Passarella's "The Monetary Circuit in the age of Financialisation" demonstrates that "return on capital" can be generated from purely financial flows rather than the functional product of real capital. For instance profits of Commercial Banks:

$$
F_b = r_{l,-1} \cdot L_{f,-1} - r_{m,-1} \cdot M_{d,-1} \quad (16)
$$

and profits of Other Financial Institutions (OFIs):

$$
F_o = r_{l,-1} \cdot L_{r,-1} + r_{c,-1} \cdot L_{w,-1} + r_{m,-1} \cdot M_{o,-1} - r_{d,-1} \cdot D_{s,-1} \quad (21)
$$



Capital makes money all sorts of ways, to the degree that it seems lately that workers are almost an afterthought. In the Uber model, for instance, big capital seeks out tiny little specks of capital to drink dry and leave the exogenous human impacts for governments to figure out.
Nowhere in Professor Mankiwʻs model is there a bank, an abstract financial instrument, or a credit card; as if these are not somehow vectors for rates of return in our modern economy. Nowhere is there an equation where r appears factored on both sides of the equation. Perhaps we are to understand that financialization is not relevant to the case at hand; but, itʻs hard to see how a discussion of national capital appreciation and growth rates would obviate the entire financial sector.
Mankiw goes on to suggest an optimum capital tax rate of exactly zero. Even if it were practical in theory to set such a tax rate---presumably buffered by national collective nightly prayers that the wealthy wonʻt remember to hide all of their income in returns---the calculation of this optimum suffers the same flaw as before. The very equations that "prove" that the availability of capital benefits all workers fail to account for instances when capital bypasses them altogether. Sometimes capital takes a spin around town without the workers at all; though not necessarily without the Prius that Mom bought.
I am somewhat suspicious that if we welded a monetary model onto Mankiwʻs "real" (M-C-M) economy we would find that he has taken us three-fourths of the way to proving Piketty correct: r > g -> spiraling income equality.

The rest of the article is politics. The prose, in todayʻs context, is surprisingly reminiscent of the back-pages musings of any given Samuel Alito Supreme Court opinion. Cruising through this editorial section we run into the particularly eye-catching:

> Yet that ideal conflicts with another—the freedom of parents to use their resources to help their children.

Help their children do what, one might ask.
